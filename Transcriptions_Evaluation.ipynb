# Swahili Transcription

## The response contains a world-level timestamp. From that, we will extract only words for further evaluations. We store these words in the predicted_transcription variable.

```python
import requests

headers = {
    'x-gladia-key': 'b76c60ca-b758-459a-a81c-ddcb881fc967',
}

files = {
    'audio_url': 'https://github.com/DriveMyScream/sample/raw/main/Swahili_Language.wav',
    'toggle_diarization': 'true',
    'language': 'swahili',
    'toggle_noise_reduction': 'true',
#    'transcription_hint': ("mkali", "kivuli", "usongeshi")
}

response = requests.post('https://api.gladia.io/audio/text/audio-transcription/', headers=headers, files=files)
```

```python
response = response.json()
gladia_transcription = []
for i in range(len(response['prediction'])):
    dict_of_words = response['prediction'][i]
    open_dict = dict_of_words['words']
    for i in range (len(open_dict)):
        word = open_dict[i]['word'].strip().lower()
        gladia_transcription.append(word)

print("Total Number of transcripted words in list:", len(gladia_transcription))        
print(gladia_transcription)
```

## 1) Swahili (Kenya)

```python
from google.oauth2 import service_account
from google.cloud import speech
```

```python
trasncription = []

client_file = 'meetminutesbackend.json'
credentials = service_account.Credentials.from_service_account_file(client_file)
client = speech. SpeechClient(credentials=credentials)

gcs_uri= 'gs://meetminutes/Swahili_Language.wav'
config = speech.RecognitionConfig(
    encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
    sample_rate_hertz=44100, 
    language_code='sw-KE'
    )
audio = speech.RecognitionAudio(uri=gcs_uri)
operation = client.long_running_recognize(config=config, audio=audio)

print("Waiting for operation to complete...")
response = operation.result(timeout=120)

for result in response.results:
    trasncription.append(result.alternatives[0].transcript)
```

```python
gcp_trasncription = []
for sentence in trasncription:
    words = sentence.split(" ")
    gcp_trasncription.extend(words)
    
print("Total Number of transcripted words in list:", len(gcp_trasncription))        
print(gcp_trasncription)
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

bleu_score_1 = sentence_bleu([' '.join(gladia_transcription)], ' '.join(gcp_trasncription),
                          weights=(1, 0, 0, 0),
                          smoothing_function=SmoothingFunction().method4)

bleu_score_2 = sentence_bleu([' '.join(gladia_transcription)], ' '.join(gcp_trasncription),
                          weights=(0.5, 0.5, 0, 0),
                          smoothing_function=SmoothingFunction().method4)

bleu_score_3 = sentence_bleu([' '.join(gladia_transcription)], ' '.join(gcp_trasncription),
                          weights=(0.33, 0.33, 0.33, 0),
                          smoothing_function=SmoothingFunction().method4)

bleu_score_4 = sentence_bleu([' '.join(gladia_transcription)], ' '.join(gcp_trasncription),
                          weights=(0.25, 0.25, 0.25, 0.25),
                          smoothing_function=SmoothingFunction().method4)

print(f'BLEU Score 1: {bleu_score_1}')
print(f'BLEU Score 2: {bleu_score_2}')
print(f'BLEU Score 3: {bleu_score_3}')
print(f'BLEU Score 4: {bleu_score_4}')
```
